```python
#loading real online data(csv file)
record=pd.read_csv("Titanic-Dataset.csv",index_col=0)  #index_col=0 
#: when we want any column from the data itself to become index instead of pandas creating 
# deafult index 
print(record)
```

```python
#methods

print(record.shape)   # (rows, columns)
print(record.columns)   # column names
print(record.index)     #index numbers
print(record.info)     # data types, non-null counts, memory usage
print(record.describe)    # count, mean, std, min, max, percentiles - numeric columns only
print(record.describe(include="all"))  # includes non-numeric columns too

print(record.size)
```

```python
print(record.isnull())  #return true if value missing
print(record.isnull().sum())  #return sum of empty values
print(record.isnull().sum().sort_values(ascending=False))  #worst offenders(most missing values)
                                                           #show at top
```
