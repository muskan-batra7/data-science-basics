```python
import pandas as pd
```

```python
record= pd.read_csv("Titanic-Dataset.csv")
record.head()
```

```python
#selecting columns
record["Age"]          #selecting 1 column-return series
record[["Age","Sex","Cabin"]]   #selecting multiple columns -return data frame
```

```python
#Acessing
record.Name
record["Name"][6]
```

```python
#.loc-label based selection
#select data based on its data index value
#.loc[rows,columns]
record.loc[0]
record.loc[0:5] #returns rows with index 0,1,2,3,4,5(5 included)
record.loc[0:6,["Name","Age","Survived"]] 
#returns row with 0,1,2,3,4,5,6 index and name,age,survived columns only
```

```python
#.iloc-index based selection
#select data based on its position
#.iloc[rows,column]
record.iloc[0]
record.iloc[0:5]  #return first 5 rows(index-0,1,2,3,4)
record.iloc[0:5,0:5]  #returns first 5 rows and only first 5 columns
record.iloc[[0,5,8],1]  #selects rows-0,5,8 and column at second position(1 index)
record.iloc[-5:]
record.iloc[-1]
```

```python
#filtering rows by condition

# passengers older than 30
record[record["Age"]>70]
# passengers in 1st class
record[record["Pclass"]==1]
# combine conditions: 1st class AND survived
record[(record["Pclass"]==1) &( record["Survived"]==1)]
# 1st class OR 2nd class
record[(record["Pclass"]==1) | (record["Pclass"]==2)]
# female passengers under 18
record[(record["Sex"]=="female") & (record["Age"]<18)]
```

```python
#combining loc with condition

record.loc[record["Age"]>60,["Name","Sex","Pclass","Age"]] 
#filter rows by age>60 and only name ,sex,pclass,age columns will be printed

record.loc[record["Pclass"]==2,["Name","Age","Pclass"]]

# count how many matched
len(record[record["Pclass"]==1])

```
